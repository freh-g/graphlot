```python
import graphlot as gv
```

```python
import random


random.seed(123127844)

grafo = gv.CreateNetworkFromRandomClasses([20,20,30], 150)
gv.plot_degree_distribution(grafo,save_fig='../images/degree_distribution.jpg')
```

```python
random.seed(123127844)
gv.visualize_network(grafo,mode = '2d',node_color_attribute='Type',legend = 'Type',cmap='viridis',annotate = 'Name' , save = '../images/network.jpg')
```

```python
random.seed(123127844)
gv.visualize_network(grafo,mode = '3d',node_color_attribute='Type',legend = 'Type',cmap='viridis',annotate = 'Name' , save = '../images/network.html')
```
